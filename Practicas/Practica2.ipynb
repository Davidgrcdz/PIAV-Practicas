```python
# Transformaciones geométricas 2D
import cv2 as cv
import numpy as np

# Variables globales
tx, ty = 0, 0
angle = 0
scale = 1.0
size = (1000, 1000)
image = cv.imread('images/cats.png', cv.IMREAD_COLOR)

def update_image():
    global tx, ty, angle, scale
    # Rotación
    angle_rad = angle * np.pi / 180
    R = np.float32([[np.cos(angle_rad), -np.sin(angle_rad)],
                    [np.sin(angle_rad), np.cos(angle_rad)]])
    
    # Traslación
    T = np.float32([[1, 0, tx],
                    [0, 1, ty]])
    
    # Similaridad (Escalado)
    S = np.float32([[scale, 0],
                    [0, scale]])
    
    # Combinación de transformaciones
    M = S @ R @ T
    transformed_image = cv.warpAffine(image, M, size)
    cv.imshow('Transformaciones', transformed_image)

def on_trackbar_rotation(val):
    global angle
    angle = val
    update_image()

def on_trackbar_translation(val):
    global tx, ty
    tx = val
    ty = val
    update_image()

def on_trackbar_resize(val):
    global scale
    scale = val / 100
    update_image()

cv.namedWindow('Transformaciones')
cv.createTrackbar('Rotation', 'Transformaciones', 0, 360, on_trackbar_rotation)
cv.createTrackbar('Translation', 'Transformaciones', 0, 200, on_trackbar_translation)
cv.createTrackbar('Resize', 'Transformaciones', 0, 200, on_trackbar_resize)

# Inicializar la imagen con valores predeterminados
update_image()

cv.waitKey(0)
cv.destroyAllWindows()
```

```python

image = cv.imread('images/cats.png')
size = (image.shape[1], image.shape[0])

# traslación
tx,ty = 100.0, 50.0
T = np.float32([[1,0,tx],
                [0,1,ty]])

imageT = cv.warpAffine(image, T, size)

cv.imshow('Traslacion', imageT)

# rotación
a = np.pi/6
T = np.float32(
    [[np.cos(a),-np.sin(a),tx],
     [np.sin(a),np.cos(a),ty]]
)
imageR = cv.warpAffine(image, T, size)

cv.imshow('Rotacion', imageR)

# similaridad
s=0.5
T = np.float32(
    [[s*np.cos(a),-s*np.sin(a),tx],
     [s*np.sin(a),s*np.cos(a),ty]])

imageS = cv.warpAffine(image, T, size)

cv.imshow('Similaridad', imageS)

cv.waitKey(0)
cv.destroyAllWindows()
```
